```python
import pandas as pd
df=pd.read_csv('healthandfitness.csv')
```

```python
df.head()
df.tail()
```

```python
df['Gender'].replace(['Male','Female'],[0,1],inplace=True)
df['Health Status'].replace(['Yes','No'],[1,0],inplace=True)
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X=df.drop(columns=['Timestamp','Name','Age in Years','Weight ','Health Status'])
```

```python
X
```

```python
y=df.drop(columns=['Timestamp','Name','Age in Years','Weight ','Gender','Heartbeat ','Systolic','Diastolic','Hemoglobin '])
```

```python
y
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=.30)
df_tree=DecisionTreeClassifier()
my_model=df_tree.fit(X_train,y_train)
df_tree
```

```python
test=df_tree.predict(X_test)
y_test
```

```python
X_test
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,test)
```

```python
accuracy
```
